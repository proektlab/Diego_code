```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as st
import scipy.io as sio
```

```python
all_flux = pd.read_csv('all_fluctuations.csv', sep=',', header=None)
all_flux_np = all_flux.to_numpy()
all_boots = pd.read_csv('all_bootstrap_var.csv', sep=',', header=None)
all_boots_np = all_boots.to_numpy()

all_flux.columns = ['PSI Baseline', 'PSI Low Ketamine', 'PSI High Ketamine', 'iCoh Baseline', 'iCoh Low Ketamine', 'iCoh High Ketamine']
all_boots.columns = ['PSI Baseline', 'PSI Low Ketamine', 'PSI High Ketamine', 'iCoh Baseline', 'iCoh Low Ketamine', 'iCoh High Ketamine']
```

```python
sns.set()
sns.set(font_scale=1.5)
```

```python
color_set = ['thistle', 'plum', 'hotpink', 'lightblue', 'lightskyblue', 'deepskyblue']
```

# Plot Fluctuation Data

```python
plt.figure()
plt.subplot(121)
sns.histplot(all_flux_np[:,0], color='deepskyblue', bins = 10)
sns.histplot(all_flux_np[:,1], color='coral', bins = 10)
sns.histplot(all_flux_np[:,2], color='pink', bins = 10)
plt.title('PSI Fluctuations - Histogram')
plt.xlabel('Magnitude of Fluctuations')
figure_size = plt.gcf().get_size_inches()
factor = 3
plt.gcf().set_size_inches(factor * figure_size)
plt.legend(['Baseline', '0.2 μg/mL Ketamine', '0.4 μg/mL Ketamine'])
plt.subplot(122)
sns.distplot(all_flux_np[:,0], color='deepskyblue', hist=False, kde_kws={'shade': True})
sns.distplot(all_flux_np[:,1], color='coral', hist=False, kde_kws={'shade': True})
sns.distplot(all_flux_np[:,2], color='hotpink', hist=False, kde_kws={'shade': True})
plt.title('PSI Fluctuations - PDF')
plt.xlabel('Magnitude of Fluctuations')
figure_size = plt.gcf().get_size_inches()
factor = 1
plt.gcf().set_size_inches(factor * figure_size)
plt.legend(['Baseline', '0.2 μg/mL Ketamine', '0.4 μg/mL Ketamine'])
plt.savefig('Fluctuation_Seaborn_Plots\PSI_Fluctuation_Desnity.png')
```

# Plot Bootstrapped Variance Distributions

```python
c = 95
b1 = ((100-c)/2)
b2 = c+((100-c)/2)
conf_int = lambda x: [np.percentile(x,b1), np.percentile(x,b2)]
all_CI = list()
all_CI.append(conf_int(all_boots_np[:,0]))
all_CI.append(conf_int(all_boots_np[:,1]))
all_CI.append(conf_int(all_boots_np[:,2]))

all_CI.append(conf_int(all_boots_np[:,3]))
all_CI.append(conf_int(all_boots_np[:,4]))
all_CI.append(conf_int(all_boots_np[:,5]))
#all_CI
```

```python
plt.figure()
sns.distplot(all_boots_np[:,0], color='skyblue')
sns.distplot(all_boots_np[:,1], color='coral')
sns.distplot(all_boots_np[:,2], color='plum')
plt.title('PSI Variance Bootstraps')
plt.xlabel('Bootstrapped Variance of Fluctuations')
figure_size = plt.gcf().get_size_inches()
factor = 3
plt.gcf().set_size_inches(factor * figure_size)
plt.legend(['Baseline', '0.2 μg/mL Ketamine', '0.4 μg/mL Ketamine'])
plt.axvline(all_CI[0][0], color='b')
plt.axvline(all_CI[0][1], color='b')
plt.axvline(all_CI[1][0], color='r')
plt.axvline(all_CI[1][1], color='r')
plt.axvline(all_CI[2][0], color='darkviolet')
plt.axvline(all_CI[2][1], color='darkviolet')
plt.savefig('Fluctuation_Seaborn_Plots\PSI_Variance_Bootstrap.png')
```

```python
plt.figure()
sns.distplot(all_boots_np[:,3], color='skyblue')
sns.distplot(all_boots_np[:,4], color='coral')
sns.distplot(all_boots_np[:,5], color='plum')
plt.title('iCoh Variance Bootstraps')
plt.xlabel('Bootstrapped Variance of Fluctuations')
figure_size = plt.gcf().get_size_inches()
factor = 3
plt.gcf().set_size_inches(factor * figure_size)
plt.legend(['Baseline', '0.2 μg/mL Ketamine', '0.4 μg/mL Ketamine'])
plt.axvline(all_CI[3][0], color='b')
plt.axvline(all_CI[3][1], color='b')
plt.axvline(all_CI[4][0], color='r')
plt.axvline(all_CI[4][1], color='r')
plt.axvline(all_CI[5][0], color='darkviolet')
plt.axvline(all_CI[5][1], color='darkviolet')
plt.savefig('Fluctuation_Seaborn_Plots\iCoh_Variance_Bootstrap.png')
```
